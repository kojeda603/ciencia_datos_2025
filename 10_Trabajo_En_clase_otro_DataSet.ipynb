```python
import pandas as pd
import os

# Ruta al directorio del dataset
dataset_dir = "/kaggle/input/life-style-data"

# Listar archivos dentro del directorio
print("Archivos en el dataset:", os.listdir(dataset_dir))

# Cargar el archivo CSV correcto
file_path = os.path.join(dataset_dir, "Final_data.csv")

# Cargar el CSV
data = pd.read_csv(file_path)

# Revisar las primeras filas y las columnas
print(data.head())
print(data.columns)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import trim_mean
from statsmodels import robust
import seaborn as sns
import matplotlib.pyplot as plt
```

##Pregunta problema:
Que diferencias hay en el gasto calorico entre mujeres y hombres?


```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Gender', y='Calories_Burned', data=data, palette='Set2')
plt.title('Distribución de Calorías Quemadas por Género')
plt.show()

print("El gráfico muestra que no hay diferencias significativas en las calorías quemadas entre hombres y mujeres. Ambos presentan medianas similares (~1250 calorías) y una dispersión comparable. Se observan varios valores atípicos altos (>2500 calorías), lo que indica algunos casos de alto gasto energético.")
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Calories_Burned', hue='Gender', data=data, alpha=0.6)
plt.title('Relación entre Edad y Calorías Quemadas por Género')
plt.show()

print("El gráfico muestra que no existe una relación clara entre la edad y las calorías quemadas. Las personas de todas las edades presentan una amplia variación en el gasto calórico, y tanto hombres como mujeres muestran patrones similares sin una tendencia evidente.")
```

```python
sns.barplot(x='Workout_Type', y='Calories_Burned', data=data, estimator='mean')
plt.title("Promedio de Calorías Quemadas por Tipo de Entrenamiento")
plt.xticks(rotation=45)
plt.show()

print("El gráfico muestra que el tipo de entrenamiento influye en las calorías quemadas. El HIIT presenta el mayor promedio (~1650 cal), seguido de Strength (~1350 cal), Cardio (~1200 cal) y Yoga (~900 cal), indicando que las rutinas de alta intensidad generan un mayor gasto energético.")
```

```python
# Duración media de la sesión de entrenamiento por tipo de ejercicio
sns.boxplot(x='Workout_Type', y='Session_Duration (hours)', data=data)
plt.title("Duración de la Sesión de Entrenamiento por Tipo de Ejercicio")
plt.xticks(rotation=45)
plt.show()
print("No existe una diferencia entre los tipos de entrenamiento")
```

```python
# Relación entre BMI y calorías quemadas
sns.scatterplot(x='BMI', y='Calories_Burned', hue='Gender', data=data)
plt.title("Relación entre IMC y Calorías Quemadas")
plt.show()
print("Existe mayor cantidad de calorias quemadas si la persina tiene un IMC entre 20 y 35 años")
```

#Respuesta a la pregunta problema
No existe una diferencia significativa en el gasto calórico entre hombres y mujeres, ya que los datos muestran un comportamiento muy similar. La variación observada se relaciona más con el tipo de entrenamiento realizado que con el género. Por ejemplo, en el caso del entrenamiento HIIT, se evidencia un mayor gasto calórico, independientemente de si es realizado por hombres o mujeres.
